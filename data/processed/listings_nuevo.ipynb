# 📊 Dataset `listings_nuevo.csv` - Documentación de Columnas

## 🎯 Descripción General

Dataset consolidado de Airbnb para análisis de turismo urbano en España, conteniendo **61,114 registros** de Barcelona, Madrid y Mallorca. Este archivo representa la versión limpia y procesada de los datos originales, optimizada para análisis avanzados.

---

## 📋 Estructura del Dataset

**Dimensiones**: 61,114 filas × 24 columnas  
**Ciudades**: Barcelona, Madrid, Mallorca  
**Fuente**: Datos originales de Airbnb procesados y consolidados  
**Fecha de creación**: Junio 2025

---

## 🔢 Columnas del Dataset

### **1. Identificadores Únicos**

| Columna | Tipo | Descripción |
|---------|------|-------------|
| `id` | int64 | **ID único del listing** - Identificador principal de cada propiedad en Airbnb |
| `host_id` | int64 | **ID del anfitrión** - Identificador único del propietario/gestor de la propiedad |

### **2. Información del Anfitrión**

| Columna | Tipo | Descripción |
|---------|------|-------------|
| `host_listings` | int64 | **Número de propiedades del anfitrión** - Cantidad total de listings que maneja |
| `superhost` | bool | **Estado de Superhost** - Indica si el anfitrión tiene certificación Superhost |

### **3. Localización Geográfica**

| Columna | Tipo | Descripción |
|---------|------|-------------|
| `latitude` | float64 | **Latitud** - Coordenada geográfica (precisión ~100m por privacidad) |
| `longitude` | float64 | **Longitud** - Coordenada geográfica (precisión ~100m por privacidad) |
| `neighbourhood_group` | object | **Distrito/Zona** - Agrupación geográfica por distritos administrativos |
| `city` | object | **Ciudad** - Barcelona, Madrid o Mallorca |

### **4. Características de la Propiedad**

| Columna | Tipo | Descripción |
|---------|------|-------------|
| `property_type` | object | **Tipo de propiedad** - Apartamento, casa, habitación, etc. |
| `room_type` | object | **Tipo de habitación** - Entire home/apt, Private room, Shared room |
| `accommodates` | int64 | **Capacidad** - Número máximo de huéspedes que puede alojar |
| `license` | object | **Licencia turística** - Número de registro oficial (puede ser nulo) |

### **5. Políticas de Reserva**

| Columna | Tipo | Descripción |
|---------|------|-------------|
| `instant_bookable` | bool | **Reserva instantánea** - Permite reserva inmediata sin aprobación |
| `minimum_nights` | int64 | **Estancia mínima** - Número mínimo de noches requeridas |
| `maximum_nights` | int64 | **Estancia máxima** - Número máximo de noches permitidas |

### **6. Disponibilidad y Calendario**

| Columna | Tipo | Descripción |
|---------|------|-------------|
| `availability_365` | int64 | **Días disponibles/año** - Número de días disponibles en los próximos 365 |
| `availability_rate_calendar` | float64 | **Tasa de disponibilidad** - Porcentaje de disponibilidad en el calendario |
| `availability_summer` | float64 | **Disponibilidad en verano** - Porcentaje de disponibilidad junio-agosto |
| `availability_weekend` | float64 | **Disponibilidad fines de semana** - Porcentaje de disponibilidad sáb-dom |

### **7. Información de Precios** 💰

| Columna | Tipo | Descripción |
|---------|------|-------------|
| `price` | float64 | **Precio por noche** - Precio base en USD (datos raw recuperados) |
| `price_volatility` | float64 | **Volatilidad del precio** - Variabilidad de precios a lo largo del tiempo |
| `summer_premium` | float64 | **Prima de verano** - Sobreprecio durante temporada alta |

### **8. Métricas Avanzadas de Turismo** 📈

| Columna | Tipo | Descripción |
|---------|------|-------------|
| `tourism_intensity` | float64 | **Intensidad turística** - Métrica calculada que mide el impacto turístico |
| `seasonal_ratio` | float64 | **Ratio estacional** - Relación entre demanda alta/baja temporada |

---

## 🔍 Notas Importantes

### **Calidad de Datos** ✅
- ✅ **Precios recuperados** desde fuentes raw (sin outliers artificiales)
- ✅ **Coordenadas validadas** dentro de rangos geográficos correctos  
- ✅ **IDs únicos** verificados sin duplicados
- ✅ **Consistencia geográfica** mantenida entre ciudades

### **Consideraciones para Análisis** ⚠️
- **Minimum_nights > 365**: Posibles propiedades de alquiler a largo plazo
- **Availability_365 = 0**: Propiedades sin disponibilidad (reservadas/inactivas)
- **License nulos**: No todas las ciudades requieren licencia turística
- **Coordenadas**: Desplazadas ~100m por políticas de privacidad de Airbnb

### **Casos de Uso Recomendados** 🎯
1. **Análisis de precios** por zona y tipo de propiedad
2. **Estudios de intensidad turística** por barrios
3. **Análisis estacional** de disponibilidad y precios
4. **Segmentación de mercado** por características de anfitrión
5. **Mapas interactivos** de distribución geográfica

```python
# 📊 ESTADÍSTICAS RÁPIDAS DEL DATASET
import pandas as pd

# Cargar dataset
df = pd.read_csv('listings_nuevo.csv')

print("📊 ESTADÍSTICAS GENERALES")
print("=" * 40)
print(f"📋 Total registros: {len(df):,}")
print(f"📊 Total columnas: {len(df.columns)}")
print(f"🌍 Ciudades: {', '.join(df['city'].unique())}")

print(f"\n🏙️ DISTRIBUCIÓN POR CIUDAD:")
city_counts = df['city'].value_counts()
for city, count in city_counts.items():
    percentage = (count / len(df)) * 100
    print(f"  • {city:9}: {count:5,} registros ({percentage:4.1f}%)")

print(f"\n🏠 TIPOS DE HABITACIÓN:")
room_types = df['room_type'].value_counts()
for room_type, count in room_types.items():
    percentage = (count / len(df)) * 100
    print(f"  • {room_type:15}: {count:5,} registros ({percentage:4.1f}%)")

print(f"\n💰 ESTADÍSTICAS DE PRECIOS:")
price_stats = df['price'].describe()
print(f"  • Precio promedio: ${price_stats['mean']:.2f}")
print(f"  • Precio mediano: ${price_stats['50%']:.2f}")
print(f"  • Rango de precios: ${price_stats['min']:.2f} - ${price_stats['max']:.2f}")

print(f"\n📅 ESTANCIAS MÍNIMAS:")
nights_stats = df['minimum_nights'].describe()
print(f"  • Promedio: {nights_stats['mean']:.1f} noches")
print(f"  • Mediana: {nights_stats['50%']:.0f} noches")
print(f"  • Máximo: {nights_stats['max']:.0f} noches")

print(f"\n🔍 CALIDAD DE DATOS:")
print(f"  • Valores nulos totales: {df.isnull().sum().sum():,}")
print(f"  • Columnas con nulos: {(df.isnull().sum() > 0).sum()}")
print(f"  • Completitud promedio: {((df.notna().sum().sum()) / (len(df) * len(df.columns)) * 100):.1f}%")
```
