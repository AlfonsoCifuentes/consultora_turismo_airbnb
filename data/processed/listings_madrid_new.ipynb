## El dataset usa **`price_market`** como precio principal, que combina: ##

1. **Precio base** → `price` del listings original (limpiado)
2. **Precio real** → `avg_price_calendar` desde 9.2M registros de calendar1
3. **Fallback** → Si no hay datos de calendar1, usa el precio de listings

## **📊 Estructura del Dataset (25 Columnas)**

### **🔍 IDENTIFICACIÓN Y CONTROL (3)**
- `id` - Identificador único del alojamiento
- `host_id` - Identificador único del anfitrión  
- `host_listings` - Número total de propiedades del host

### **📍 UBICACIÓN (3)**
- `latitude` - Coordenada de latitud
- `longitude` - Coordenada de longitud
- `neighbourhood_group` - Distrito o zona de Madrid

### **🏠 TIPOLOGÍA DEL ALOJAMIENTO (3)**
- `property_type` - Tipo de propiedad (Apartamento, Casa, etc.)
- `room_type` - Modalidad (Entire home/apt, Private room, etc.)
- `accommodates` - Capacidad máxima de huéspedes

### **⚖️ REGULACIÓN Y POLÍTICAS (5)**
- `license` - Número de licencia turística oficial
- `superhost` - Estatus de "Superhost" (True/False)
- `instant_bookable` - Reserva instantánea habilitada
- `minimum_nights` - Estancia mínima requerida
- `maximum_nights` - Estancia máxima permitida

## **+ CALENDAR1.CSV**

### **📅 DISPONIBILIDAD TEMPORAL (4)**
- `availability_365` - Días disponibles año próximo (listings base)
- `availability_rate_calendar` - **Tasa real de disponibilidad** (desde calendar1)
- `availability_summer` - **Disponibilidad estival** (Jun-Sep desde calendar1)
- `availability_weekend` - **Disponibilidad fines de semana** (desde calendar1)
  

### **💰 PRECIOS TEMPORALES (5)**
- `price_market` - **Precio consolidado** (calendar1 + fallback listings)
- `avg_price_summer` - **Precio promedio verano** (desde calendar1)
- `price_weekend` - **Precio promedio fin de semana** (desde calendar1)
- `price_volatility` - **Volatilidad de precios** (desviación estándar desde calendar1)
- `summer_premium` - **% Premium estacional** (verano vs resto del año)

### **🎯 MÉTRICAS CONSOLIDADAS (2)**
- `tourism_intensity` - **Score 0-100 de intensidad turística** (combina premiums temporales)
- `seasonal_ratio` - **Ratio estacional** (verano/invierno-primavera)
